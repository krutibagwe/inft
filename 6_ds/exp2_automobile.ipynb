# **Automobile Dataset**

```python
import pandas as pd
import io

df = pd.read_csv("/content/Automobile_data.csv")
df.info()
```

```python
#Barplot
import numpy as np
import matplotlib.pyplot as plt

x=df['body-style'].value_counts().sort_index().index
y=df['body-style'].value_counts().sort_index().values
plt.bar(x,y,color='green')
plt.title('Body style frequency')
plt.xlabel("Body style")
plt.ylabel("Frequency")
```

```python
import matplotlib.pyplot as plt
df['price'] = pd.to_numeric(df['price'])
clean_data = df.dropna(subset=['price'])
body_style_avg_price = clean_data.groupby('body-style')['price'].mean()
body_style_std_price = clean_data.groupby('body-style')['price'].std()
body_styles = body_style_avg_price.index
avg_prices = body_style_avg_price.values
std_prices = body_style_std_price.values
plt.bar(body_styles, avg_prices, yerr=std_prices, capsize=5, color='purple', alpha=0.7)
plt.title("Average Price by Body Style with Error Bars")
plt.xlabel("Body Style")
plt.ylabel("Average Price (USD)")
plt.xticks(rotation=45)
plt.show()
```

```python


import matplotlib.pyplot as plt
#histogram
plt.hist(df['city_mpg'], bins=10, alpha=0.5, color='purple', label='City MPG')
plt.legend(loc='upper right')
plt.title('City MPG Distribution')
plt.xlabel('City MPG')
plt.ylabel('Frequency')
plt.grid(True, color='grey', linestyle='--', linewidth=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
#pie chart
fuel_count = df['fuel'].value_counts()
plt.pie(fuel_count, labels=fuel_count.index, autopct='%1.3f%%', startangle=90)
plt.title('Distribution of Fuel type')
plt.show()
```

```python





x = df['num-of-cylinders'].value_counts().sort_index().index
y = df['num-of-cylinders'].value_counts().sort_index().values
plt.plot(x, y)
plt.title("Number of Cylinders Distribution")
plt.xlabel("Number of Cylinders")
plt.ylabel("Count")
plt.show()
```

```python
x1=df['city_mpg'].value_counts().sort_index().index
y1=df['city_mpg'].value_counts().sort_index().values
x2=df['highway_mpg'].value_counts().sort_index().index
y2=df['highway_mpg'].value_counts().sort_index().values
plt.title("City MPG vs Highway MPG")
plt.xlabel("City MPG")
plt.ylabel("Highway MPG")
plt.plot(x1, y1, label='City MPG', color='blue')
plt.plot(x2, y2, label='Highway MPG', color='green')
plt.legend(loc='upper right')
```

```python

#scatterplot
x = df.highway_mpg
y = df.length
plt.scatter(x, y)
plt.title("Highway MPG vs Length")
plt.xlabel("Highway MPG")
plt.ylabel("Length")
```

```python
x_city_mpg = df['city_mpg']
x_highway_mpg = df['highway_mpg']
y = df['length']
plt.scatter(x_city_mpg, y, color='blue', label='City MPG')
plt.scatter(x_highway_mpg, y, color='orange', label='Highway MPG')
plt.title("MPG vs Length (City MPG in Blue, Highway MPG in Orange)")
plt.xlabel("MPG")
plt.ylabel("Length")
plt.legend()
plt.show()
```
